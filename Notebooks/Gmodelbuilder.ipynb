```python
# handle rare category function from Tran Minh Nhat, my teammate code suggestion
def handle_rare_categories(df, column, threshold=0.05):
    value_counts = Gas[column].value_counts(normalize=True)
    rare_categories = value_counts[value_counts < threshold].index
    Gas[column] = Gas[column].apply(lambda x: 'Other' if x in rare_categories else x)
    return Gas

Gas = handle_rare_categories(Gas, 'net_manager', threshold=0.05)
Gas = handle_rare_categories(Gas, 'purchase_area', threshold=0.05)
```

```python
# display outliers
def find_outliers(data):
    # Calculate the interquartile range (IQR)
    for col in list(data.columns):
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1

        # Calculate the lower and upper bounds for outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Count the number of outliers
        num_outliers = len(data[(data[col] < lower_bound) | (data[col] > upper_bound)])
        print(f"{col} has {num_outliers} outliers")
    
find_outliers(Gas_numeric)
```
